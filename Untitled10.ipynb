```python
import random
```

```python
correct_password = "admin123"
max_attempts = 5
attempts = 0
```

```python
print("secure login system started")
```

```python
while attempts < max_attempts:
    password = input("enter your password: ")
    if password == correct_password:
        print("login successful")
        break
    else:
        attempts += 1
        print("incorrect password")
        if attempts >= 3:
            a = random.randint(1,10)
            b = random.randint(1,10)
            captcha_input = int(input(f"\nCAPTCHA Verification required: What is {a} + {b}? "))

            if captcha_input != a + b:
                print("CAPTCHA Failed. Suspicious Activity Detected")
                continue
        else:
            print(f"Remaining attempts: {max_attempts - attempts}")
if attempts == max_attempts:
  print("\n IDS ALERT: possible brute force attack detected")
  print("Account locked")
```
