```python
# Brute Force Attack Detection using IDS

correct_password = "admin123"
max_attempts = 5
attempts = 0

print("===== Login Monitoring System (IDS) =====")

while attempts < max_attempts:
    password = input("Enter password: ")

    if password == correct_password:
        print("Login Successful")
        break
    else:
        attempts += 1
        print("Incorrect Password")

        remaining = max_attempts - attempts
        print("Attempts remaining:", remaining)

if attempts == max_attempts:
    print("\nALERT: Possible Brute Force Attack Detected!")
    print("Account Locked by IDS.")
```

```python
correct_password = "1a&@A"
password_list = ["1111", "2222", "3333", "1234", "0000", "5555", "1234@"]

max_attempts = 5
attempts = 0
password_found = False

print("=== IDS Monitoring Login Attempts ===")

for password in password_list:

    if attempts >= max_attempts:
        print("\n⚠ ALERT: Possible Brute Force Attack Detected!")
        print("Maximum threshold (5 attempts) exceeded.")
        break

    attempts += 1
    print("Attempt", attempts, "- Trying password:", password)

    if password == correct_password:
        print("✅ Password found:", password)
        password_found = True
        break
    else:
        print("❌ Incorrect password")

if not password_found and attempts < max_attempts:
    print("\nPassword not found in provided password list.")
```

```python
from sklearn.metrics import confusion_matrix
import pandas as pd

# actual data
attempts_data = [2, 3, 7, 1, 6, 4, 8]

# ground truth
actual_labels = ["Normal","Normal","Attack","Normal","Attack","Normal","Attack"]

# IDS threshold rule
threshold = 5

predicted_labels = []

for attempts in attempts_data:
    if attempts > threshold:
        predicted_labels.append("Attack")
    else:
        predicted_labels.append("Normal")

# confusion matrix
cm = confusion_matrix(actual_labels, predicted_labels, labels=["Normal","Attack"])

df = pd.DataFrame(cm,
                  index=["Actual Normal","Actual Attack"],
                  columns=["Predicted Normal","Predicted Attack"])

print("Confusion Matrix\n")
print(df)
```
